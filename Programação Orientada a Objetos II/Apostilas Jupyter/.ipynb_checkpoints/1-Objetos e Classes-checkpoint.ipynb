# 1 - Objetos e Classes

## Python

Python  uma linguagem que contempla tanto o paradigma **imperativo** quanto o paradigma **orientado a objetos**:
- **Imperativo:** o programa  uma sequncia de comandos que modificam variveis (estados). O foco  em *como resolver o problema*, utilizando comandos condicionais, de repetio, funes, atribuies, etc, para manipular dados (variveis).
- **Orientado a Objetos:** O foco  na representao computacional do problema, identificando "objetos" importantes do problema e "organizando-os". Tais objetos mantm, dentro de si, seus prprios dados e as operaes que manipulam esses dados. O foco  em *quem est sendo afetado na resoluo do problema*. Um importante conceito de OO : **abstrao**.

---

## O que  uma classe?

Considere o nvel de abstrao da segunda figura e que tenhamos diversas livros: Harry Potter, Dom Casmurro, O Senhor dos Anis, Aprendendo Python, etc... Neste caso, o que significa a representao de livro da figura 1? Certamente no  um objeto, pois no  especfico o suficiente de acordo com a nossa abstrao, mas  um conceito mais geral, que engloba todos os objetos livros.

>*Uma classe  a definio de um grupo de objetos com **propriedades** e **comportamentos** comuns. Assim, um objeto  uma **instncia** de uma classe.*

Por exemplo, um carro tem as seguintes propriedades (alm de outras):
- modelo
- cor
- placa
- velocidade

E os seguintes comportamentos (alm de outros):
- buzinar()
- acelerar()
- frear()

Qualquer objeto carro ter as propriedades e comportamentos acima.

Podemos estender ainda mais a nossa definio de objeto:

>*Um objeto  uma representao computacional de algo, para executar uma determinada tarefa, com limites bem definidos. ** uma coleo de dados (atributos) e operaes (mtodos) para manipular estes dados, representando uma entidade lgica no sistema**.*

---
### Exemplo:

**Classe** `Carro`

>**Atributos:**
>- modelo
>- cor
>- placa
>- velocidade
>
>**Mtodos:**
>- buzinar()
>- acelerar()
>- frear()

**Objeto** `Carro 1`:
>- modelo = Gol
>- cor = Vermelho
>- placa = ABC-1234
>- velocidade = 0

**Objeto** `Carro 2`:
>- modelo = Fox
>- cor = Prata
>- placa = XYZ-1234
>- velocidade = 80

Aqui temos 2 objetos da classe `Carro`, cada um especificando os seus valores para cada atributo. Ambos os objetos podem "executar" qualquer mtodo da classe. Nesse sentido, os mtodos podem ser vistos como funes (do paradigma imperativo), mas que ficam "dentro" dos objetos, pois no fazem sentido existirem se no estiverem atrelados a estes.

---
## Atributos e Mtodos

>*Atributos so variveis que descrevem as caractersticas um objeto.*

>*Mtodos so funes que definem o comportamento de um objeto.*

---
## Declarar uma Classe em Python

```python
class Carro:
    def __init__(self, modelo: str, cor: str, placa: str):
        self.modelo = modelo
        self.cor = cor
        self.placa = placa
        self.velocidade = 0
        
    def buzinar(self):
        print('carro', self.modelo, 'buzinou')
        
    def acelerar(self):
        self.velocidade += 10
        
    def frear(self):
        self.velocidade -= 10
        if self.velocidade < 0:
            self.velocidade = 0
```

- `self:`  uma referncia ao objeto desta classe que ser instanciado ( obrigatrio e deve ser sempre o primeiro parmetro dos mtodos definidos na classe).
- `__init__():`  um mtodo especial chamado de **construtor** que  executado automaticamente sempre que a classe  instanciada.  responsvel por inicializar atributos. Pode chamar quaisquer mtodos da classe para isso.
- Observe que os atributos tm tipos especficos, que podemos descrever com a notao `atributo: tipo`. Neste exemplo, os atributos `modelo`, `cor` e `placa` da classe `Carro` so strings, enquanto o atributo `velocidade`  um inteiro.

Instanciar uma classe em Python:

```python
carro1 = Carro('Gol', 'Vermelho', 'ABC-1234')
carro2 = Carro('Fox', 'Prata', 'XYZ-1234')
    
carro1.buzinar()
carro2.acelerar()
print('velocidade do', carro2.modelo, ':', carro2.velocidade, 'km/h')
```

Note como o parmetro `self` **no ** utilizado na chamada do mtodo. Note tambm a diferena entre acessar mtodos e atributos do objeto `carro2`: para acessar os mtodos utiliza-se parnteses, diferentemente do acesso aos atributos.

Mtodos tambm podem ter argumentos e retorno.
Exemplo:

```python
class Carro:
    def __init__(self, modelo: str, cor: str, placa: str):
        self.modelo = modelo
        self.cor = cor
        self.placa = placa
        self.velocidade = 0
        
    def buzinar(self):
        print('carro', self.modelo, 'buzinou')
        
    def acelerar(self, valor: int):
        self.velocidade += valor
        if self.velocidade > 100:
            return 'cuidado!'
        else:
            return 'velocidade aceitvel.'
        
    def frear(self):
        self.velocidade -= 10
        if self.velocidade < 0:
            self.velocidade = 0
            
            
def main():
    carro1 = Carro('Gol', 'Vermelho', 'ABC-1234')
    carro2 = Carro('Fox', 'Prata', 'XYZ-1234')
    
    carro1.buzinar()
    print(carro2.acelerar(80))
    print(carro2.velocidade, 'km/h')
    
main()
```

## Exerccio

Faa um programa que leia uma string `s` indicando um tipo de forma geomtrica (considere que `s` pode ser `'quadrado'`, `'retangulo'` ou `'circulo'`). Conforme a forma geomtrica lida, seu programa deve ler tambm a(s) sua(s) dimenso(es), sendo `lado` para o quadrado, `base` e `altura` para o retngulo e `raio` para o circulo. Seu programa deve mostrar qual a rea da forma geomtrica lida. Para resolver este problema, pense quais so os objetos envolvidos e qual o nvel de abstrao adequado. O que seria um objeto nesse caso? E o que seria uma classe? Defina classes adequadas, com os respectivos atributos e mtodos, e instancie seus objetos no programa principal.

```python
import math
class FormaGeometrica:
    def __init__(self, tipo, lado=0, base=0, altura=0, raio=0):
        self.tipo = tipo
        self.lado = lado
        self.base = base
        self.altura = altura
        self.raio = raio
    
    def area(self):
        if self.tipo == "quadrado":
            return self.lado**2
        elif self.tipo == "retangulo":
            return self.base*self.altura
        else:
            return math.pi*(self.raio**2)
        
circulo1 = FormaGeometrica("circulo", raio=4)
retangulo1 = FormaGeometrica("retangulo", base=2, altura=9)
print(circulo1.area())
print(retangulo1.area())
        
```
